```
# change this to your files name when you upload it
FILE_NAME = "iris dataset - Sheet1.csv"
```

```
# import the data and view it

import pandas as pd

data = pd.read_csv(FILE_NAME)
data
```

```
# create our input data X (the predictors) and output data y (the thing we want to predict) variables

X = data.drop('variety', axis = 1)
y = data['variety']
```

```
# import desicion tree package
from sklearn import tree
```

```
# create the model , you can play around the max_depth value
model = tree.DecisionTreeClassifier(max_depth=2)

# train the model
model.fit(X, y)
```

```
# see performance
model.score(X, y)
```

```
# display decision tree

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
tree.plot_tree(model, 
               feature_names=X.columns, 
               class_names=y.unique())
plt.show()
```
